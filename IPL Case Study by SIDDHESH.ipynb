# Which player scored the most runs in the dataset?






```python
# psr = PLAYER SCORED RUNS 

psr = df.groupby('batsman')['batsman_runs'].sum()
psr
```

```python
# tsr = TOTAL SURESH RAINA RUNS

tsr = psr.idxmax()
tsr
```

```python
print(f'The Player Who Scored Most Runs in this IPL Dataset is {tsr} with {psr[tsr]} runs.')
```
